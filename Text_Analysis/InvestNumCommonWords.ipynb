```python
import pandas
import nltk
import re
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
#Do this once when you fisrt install :) 
nltk.download('all')
```

```python
Stop_Words = stopwords.words('english')
#print(Stop_Words)
New_Stop_Words = pandas.read_csv('stopwords_research.csv')
Partial_Stop_Words = list(New_Stop_Words['words'])
#print(Partial_Stop_Words)
```

```python
Articles = pandas.read_csv('articles.csv', engine='python')
Title = Articles['Title']
Abstracts = Articles['Abstract']
NumArt = len(Abstracts)
```

```python
EngWords = []
TechWords = []
Keywords = []
TotalWords = []
for i in range(NumArt):
    Line = str(Abstracts[i])
    line = Line.lower()
    #print(line)
    Num_Stop_Words  = 0
    for word in Stop_Words:
        word = r'\b'+word+r'\b'
        matches = re.findall(word,line)   
        Num_Stop_Words = Num_Stop_Words + len(matches)
        line = re.sub(r'[^\w\s]','',line)
    #print(Num_Stop_Words)
    Num_Tech_Stop_Words = 0
    for word in Partial_Stop_Words:
        word = r'\b'+ word +r'\w*.?,?\s'
        matches = re.findall(word,line) 
        Num_Tech_Stop_Words = Num_Tech_Stop_Words + len(matches)
    #print(Num_Tech_Stop_Words)
    text = line.split()
    Len_Abs = len(text)
    #print(Len_Abs)
    
    if Len_Abs > 10:
        EngWords.append(round(Num_Stop_Words*100/Len_Abs,2))
        TechWords.append(round(Num_Tech_Stop_Words*100/Len_Abs,2))
        Perc_Keywords = Len_Abs - Num_Tech_Stop_Words - Num_Stop_Words
        Keywords.append(round(Perc_Keywords*100/Len_Abs,2))
        TotalWords.append(Len_Abs)
    else:
        continue
```

```python
#print(EngWords)
#print(TechWords)
#print(Keywords)
#print(TotalWords)
print(len(EngWords))
A = len(EngWords)
```

```python
import csv
filename = 'article_Language_Style.csv'
with open(filename, mode='w') as f:
    fieldnames = ['Title', 'Total Abstract Length','Perc common English words','Perc common technical words','Perc keywords']
    writer = csv.writer(f, delimiter=',')
    writer.writerow(fieldnames)
    for i in range(A):
        List = [Title[i],TotalWords[i],EngWords[i],TechWords[i],Keywords[i]]
        try:
            writer.writerow(List)
        except:
            continue
```
